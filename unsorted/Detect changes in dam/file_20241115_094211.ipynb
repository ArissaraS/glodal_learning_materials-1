## 8.1.2 (Basic) [Reading and visualizing ALOS-2 data]
### a. (Basic) [Data format (*.tif)](https://www.bluemarblegeo.com/blog/about-geotiff-format/#:~:text=TIF%2C%20or%20TIFF%20stands%20for,image%20data%20in%20geospatial%20workflows.)

TIF, or TIFF stands for Tag Image File Format. Data in this format is pixel-based, and it is often used for high-quality graphics, making it a great fit for image data in geospatial workflows. 

GeoTIFF is a variation of the TIF format that adds a set of tags containing geospatial data in order to provide internal georeference information for the raster data in the file. This means that a single GeoTIFF can be loaded into Global Mapper and will display at the intended and geodetically correct location with no manual rectification needed.

#### Load ALOS Data Using Rasterio

You can load the ALOS data by specifying the path to the GeoTIFF file. ALOS data often comes with multiple bands (e.g., HH, HV), so we’ll load the desired band for visualization.

```
import rasterio
import matplotlib.pyplot as plt

# Path to the ALOS GeoTIFF file
alos_file = '/home/jovyan/shared/Arissara/ALOS-2/Yamba/Finalize/Yamba/IMG-HH-ALOS2310942880-200225-UBSR2.1GUD.tif'

# Open the GeoTIFF file using rasterio
with rasterio.open(alos_file) as src:
    # Read the first band (typically ALOS data has multiple bands, choose the appropriate one)
    alos_data = src.read(1)  # Read the first band (HH, HV, etc.)
    transform = src.transform  # Affine transform
    crs = src.crs  # Coordinate reference system

# Print some basic metadata
print(f"CRS: {crs}")
print(f"Transform: {transform}")
print(f"ALOS Data Shape: {alos_data.shape}")
```

#### Visualize the ALOS Data 

Once the data is loaded, you can use `matplotlib` to display the ALOS image.

```
# Visualize the ALOS data using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(alos_data, cmap='gray')
plt.title('ALOS Data (Band 1)')
plt.colorbar(label='Pixel values')
plt.show()
```

### Apply a Custom Color Map (Optional) 

You can enhance the visualization by applying a custom colormap to highlight specific features in the ALOS image.

```
# Visualize with a custom colormap (e.g., 'viridis', 'terrain', etc.)
plt.figure(figsize=(10, 10))
#plt.imshow(alos_data, cmap='viridis')
plt.imshow(alos_data, cmap='turbo')
plt.title('ALOS Data (Band 1) with Viridis Colormap')
plt.colorbar(label='Pixel values')
plt.show()
```

#### Compute the histogram of the ALOS data

```
import numpy as np

# Flatten the 2D array to 1D for histogram calculation
hist_values, bin_edges = np.histogram(alos_data, bins=256, range=(np.min(alos_data), np.max(alos_data)))

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.plot(bin_edges[:-1], hist_values, lw=2)  # bin_edges[:-1] because np.histogram returns one extra edge
plt.title('Histogram of ALOS Data (Band 1)')
plt.xlabel('Pixel values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
